2. We could combine the date and time columns to see trends of entries and exits over time

<b> Implementing:</b>

```python
# 1. Making the Columns Lower Case
data.columns = [i.lower().strip() for i in data.columns]

# 2. Creating a New date_time Column

# Joining the date and time columns
date_time = [date + ' ' + data.time.iloc[i] for i, date in enumerate(data.date)]
# Converting it to a datetime object for a new date_time columns
data['date_time'] = map(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S'), date_time)

# 3. Making the Dates into Datetime
data.date = map(lambda x: x.date(), data.date_time)
```

# <u> 4.  Feature Engineering

### A. Making Features By Segmenting Time

It will be useful to segment the entries and exits by different time intervals including months, years, hours, etc.
<br>So next up we will create some features for the different time segments.

```python
# Month
data['month'] = [i.strftime('%b') for i in data.date]

# Year
data['year'] = [i.year for i in data.date]

# Day of the Month
data['day'] = [i.day for i in data.date]

# Days of the Week
data['day_of_week'] = [calendar.day_name[i.weekday()] for i in data.date]

# Whether it's a Weekend
weekend = ['Saturday', 'Sunday']
data['weekend'] = [1 if i in weekend else 0 for i in data.day_of_week]

# The hour of the day as an integer
data['time_int'] = [int(i.split(':')[0]) for i in data.time]
```

```python
# Season
def get_season(date):
    year = date.year
    seasons = {'Summer':(datetime(year,6,1).date(), datetime(year,8,31).date()),
           'Autumn':(datetime(year,9,1).date(), datetime(year,11,30).date()),
           'Spring':(datetime(year,3,1).date(), datetime(year,5,31).date())}
            
    for season,(season_start, season_end) in seasons.items():
        if date>=season_start and date<= season_end:
            return season
    else:
        return 'Winter'
    
data['season'] = [get_season(i) for i in data.date]
```

### B. Creating Net Entries & Exits

We will calculate the net exits and entries by subtracting the entries of a later time to one row earlier.<br>

<b>Method:</b>
1. I will do this by making a list of entries from the second value to the end and subtracting that from a list
of entries from the first value to the second to last value. <br>
2. Check for any anomalies like negative value or values too large to make sense. 
3. Create a new feature with only the rows that make sense.

```python
# 1. Creating the Raw Net Value
exits = data.exits
entries = data.entries

net_exits = exits[1:].values - exits[:-1].values
net_entries = entries[1:].values - entries[:-1].values
```

```python
# Net Entries
pd.Series(net_entries).hist()
```

```python
# Net Exits
pd.Series(net_exits).hist()
```

<b>Take-aways:</b>
1. There are clearly some very large numbers for the histogram to be composed of only two bins. So we will have to
get rid of these outliers
2. There are also negative values which do not make sense and we may have to get rid of.

<b>Implementation:</b>
1. I will make the <u>assumption</u> that the turnstiles sometimes make the mistake of counting turns as negative so
I will take the absolute value of all the entries and exits
2. I will then see the new distribution and delete the outliers

```python
# 1. Creating Net Value
abs_net_exits =   np.abs(exits[1:].values - exits[:-1].values)
abs_net_entries = np.abs(entries[1:].values - entries[:-1].values)

# I will create a new dataframe with the net values
net_data = data.copy().loc[1:]
net_data['abs_net_entries'] = abs_net_entries
net_data['abs_net_exits'] =   abs_net_exits
```

```python
net_data.abs_net_entries.describe()
```

```python
net_data.abs_net_exits.describe()
```

These numbers suggest that at certain times there are an average of 2,000,000 people coming in or out of a single 
subway station which seems <u>WAY</u> to high. <br><br>
<b>I will limit the output to certain numbers and see when the distribution makes sense.

Distribution if we Limit net entries to maximum 10,000

```python
temp = net_data.abs_net_entries[net_data.abs_net_entries <10000]
print 'Percentage of Data Lost: %r '%(round(((net_data.shape[0] - temp.shape[0])/float(net_data.shape[0])),3)*100)
temp.hist()
```

Distribution if we Limit net exits to maximum 10,000


```python
temp = net_data.abs_net_exits[net_data.abs_net_exits <10000]
print 'Percentage of Data Lost: %r '%(round(((net_data.shape[0] - temp.shape[0])/float(net_data.shape[0])),3)*100)
temp.hist()
```

Distribution if we Limit net exits and entries to maximum 10,000


```python
temp = net_data.abs_net_exits[(net_data.abs_net_exits <10000)& (net_data.abs_net_entries <10000)]
print 'Percentage of Data Lost: %r '%(round(((net_data.shape[0] - temp.shape[0])/float(net_data.shape[0])),3)*100)
temp.hist()
```

<b> Takeaways:</b>
1. If we limit the data to maximimum being 10,000 entries we only lose 2.3% of the data. 
2. The fact that when we limit both extries and exits to 10,000 and we don't lose any extra rows means that
the rows with outliers in entries are also the ones with outliers in exits.
3. Removing these rows will only cause us to love 2.3% of the data

<u><i> Creating a Dataframe that Removes the Outliers

```python
final_data = net_data.copy()[(net_data.abs_net_exits <10000)& (net_data.abs_net_entries <10000)]
```

# <u> 5. Data Exploration

## A. By Time Segmentations

### <i> Weekdays

```python
days_1 = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
final_data.groupby('day_of_week').mean()['abs_net_entries'][days_1].plot()
plt.title('Entrances Per Day of Week')
plt.xlabel('Time')
plt.ylabel('Entries')
print ''
```

There is much more traffic on the workweeks with Wednesday getting the most traffic of the week. <br>
<b>So I'll see the entries first per workweek and then for the weekends.</b>

### <i> Heatmap Days of Week vs Hour

```python
days_2 = [ 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday','Sunday']
day_hour = final_data[final_data.season == 'Summer'].groupby([ 'day_of_week', 'time_int']).sum()['abs_net_entries'][days_2[::-1]].unstack()

plt.figure(figsize = (12,6))
sns.heatmap(day_hour, cmap = 'viridis')
plt.title('# of Entries Per Day and Time', size = 25 )
plt.ylabel('Day of Week',size = 17)
_ = plt.xlabel('Hour',size = 17)
```

<b> Takeaways:</b>
1. There is more subway traffic during the week.
2. There is a clear defined deadzone of subway usage from 1am to 7am.
3. The highest peak in subway usage is midweek during 8 p.m.
4. There are 3 peaks: 12p.m, 3 p.m., and 8 p.m.


<i><u>Midweek at 8p.m should be put on top considerations for street teams to work.

## B. By Station Segmentation in Relation to Time

To make recommendations depending on workweek or weekend times, we will see which stations are the most trafficked during the week and weekend and then see their in coming traffic for the top 5 stations.
